# Different experiments with TensorBoard

```python
from torch.utils.tensorboard import SummaryWriter
```

## Visualization of spectrograms with tensorboard:

```python
import preprocessing
import config
import matplotlib.pyplot as plt
```

```python
conf = config.Config(80)
conf.tb = SummaryWriter(log_dir='TensorBoardTest')
```

```python
preprocessor = preprocessing.Audio_preprocessor(conf, False)
```

```python
conf.tb.add_audio('audio_sample', preprocessor.read_audio('./data/origin_data/train_curated/7a162942.wav', True))
fig, ax = plt.subplots(figsize=(15, 5))
ax.set_yscale('linear')
ax.set_xscale('linear')
ax.imshow(preprocessor.read_as_melspectrogram('./data/origin_data/train_curated/7a162942.wav'), cmap="Spectral", interpolation='nearest')
print(type(fig))
ax.set_aspect('auto')
conf.tb.add_figure('spectrogram_sample', figure=fig)
```

```python
conf.tb.close()
```
